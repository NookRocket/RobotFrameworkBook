## Section 3 : Android automation with Robot Framework and Appium 

### Installation

* Download JDK
https://www.oracle.com/th/java/technologies/downloads/#jdk23-windows
* Download Android Studio https://developer.android.com/studio then open it to install dependent SDK
    1. create a project
    2. open SDK Manager
    3. System Settings > Android SDK
    4. Uncheck Hide Obsolete Packages
    4. Install SDK Tools
        - Android SDK Tools (Obsolete)
        - Google USB Driver
        - Google Play services
        - Intel x86 Emulator Accelerator (HAXM installer)
* Download NodeJS Package Manager https://nodejs.org/en/download/package-manager
* Install Appium Client `npm install -g appium`
* Install Appium Tool `npm install -g appium-doctor` : Appium-doctor is a command-line tool that helps diagnose and fix configuration issues for the Appium automation framework
* Install Python Appium Library `pip install robotframework-appiumlibrary`
* Set environments to 
    - ANDROID_HOME=..\AppData\Local\Android\Sdk
    - JAVA_HOME=C:\Program Files\Java\jdk[version]
    - PATH += %ANDROID_HOME%\platform-tools
* Run command `appium-doctor` to check the dependencies


#### Configure Devices And Emulators

`adb devices` to list the connected devices

Create an emulator

1. Device Manager > Add New Device
2. Choose a device. Note: choosing which supporting Play Store, it doesn't let to modify operating system because it is an official operating system then Next
3. Choose a system image then Next
4. Name the device `{device}_{operatingSystem}` then Finish
